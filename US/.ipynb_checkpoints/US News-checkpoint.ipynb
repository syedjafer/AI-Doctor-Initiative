```python
class Doctor:
    def __init__(self):
        self.doc_name = ''
        self.speciality = ''
        self.phone = ''
        self.desc = ''
        self.gender = ''
        self.hospital = ''
        self.profile_pic = ''
        self.city = ''
        self.state = ''
        self.zip_code = ''
        self.street_address = ''
        self.sub_specialities = ''
        self.years_of_experience = ''
        self.languages = ''
        self.name_prefix = ''
        self.title = ''
        
        
```

```python
def create_doc_obj(data):
    doc = Doctor()
    doc.doc_name = data.get("full_name","")
    doc.speciality = data.get("specialty", "")
    doc.phone = data.get("phone", "")
    doc.desc = data.get("blurb","")
    doc.gender = data.get("gender", "")
    doc.hospital = data.get("affiliated_hospital", "")
    doc.profile_pic = data.get("photo", "")
    location = data.get("location", {})
    doc.city = location.get("city", "")
    doc.state = location.get("state", "")
    doc.zip_code = location.get("zip_code", "")
    doc.street_address = location.get("street_address", "")
    doc.sub_specialities = data.get("subspecialties", [])
#     doc.years_of_experience = sum(data.get("years_of_experience",[]))
    doc.languages = ",".join(data.get("languages", []))
    doc.name_prefix = data.get("name_prefix", '')
    doc.title = data.get("title", "")
    return doc
```

```python
url = "https://health.usnews.com/health-care/doctors/search-data?distance=any&page_num={}"
page_count = 1
doc_list = []
total_pages = 50
while total_pages > 0:
    page = url.format(page_count)
    req = requests.get(page, headers=headers)
    if req.status_code == 200:
        all_data = req.json()
        docs = all_data["doctor_search"]["results"]['doctors']['matches']
        for data in docs:
            doc_obj = create_doc_obj(data)
            doc_list.append(doc_obj)
            print(doc_obj.doc_name)
    page_count += 1
    total_pages -= 1
    print("Remaining Pages {}".format(total_pages))
        
```
